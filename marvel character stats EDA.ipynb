```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
marvel = pd.read_csv('marvel_charcters_stats.csv')
```

```python
marvel.head()
```

```python
marvel.shape
```

```python
marvel['Alignment'].value_counts()
```

```python
good = marvel[marvel['Alignment']=='good']
```

```python
good.head()
```

```python
good.sort_values(by='Speed',ascending=False).head()
```

```python
good.sort_values(by='Power',ascending=False).head()
```

```python
good_max_power = good[good['Power']==100]
```

```python
good_max_power.shape
```

```python
good_max_power = good_max_power.sort_values(by='Total',ascending=False).head()
```

```python
plt.figure(figsize=(8,8))
plt.bar(list(good_max_power['Name'][:5]),list(good_max_power['Total'][:5]),color=['r','b','k','c','g'])
plt.show()

```
